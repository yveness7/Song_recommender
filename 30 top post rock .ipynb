```python
url = 'https://www.stereogum.com/1711293/30-essential-post-rock-songs/lists/'
from bs4 import BeautifulSoup
import pprint
import requests
response = requests.get(url)
print(response.status_code)
```

```python
soup = BeautifulSoup(response.content)
soup.select('title')
```

```python
print(soup.prettify())
```

```python
for a in soup.select('h2'):
    print(a.get_text())
```

## bands

```python
bands = []
for i in soup.select('h2')[1:31]:
    #i.split(' – ')[0]
    band_name = i.get_text().split(' – ')[0]  #we extract the text and then split
    bands.append(band_name)

bands
```

## songs

```python
songs = []
for i in soup.select('h2')[1:31]:
 
    song_info = i.get_text().split(' – ')[1]  #we extract the text and then split
    song_title = song_info.split(' (from')[0]
    songs.append(song_title)

   
    
    
songs
```
